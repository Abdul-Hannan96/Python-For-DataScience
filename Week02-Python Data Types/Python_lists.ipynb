## Creating a List

```python
# Empty
print([])
# 1D -> Homo
print([1,2,3,4,5])
# 2D -> nested list
print([1,2,3,[4,5]])
# 3D
print([[[1,2],[3,4]],[[5,6],[7,8]]])
# Hetrogenous
print([1,True,5.6,5+6j,'Hello'])
# Using Type conversion
print(list('hello'))
# string
print(list('123'))
```

## Accessing Items from a List

```python
L = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(list((L[0][0][1],L[0][1][1])))
```

```python
# Indexing
L = [[[1,2],[3,4]],[[5,6],[7,8]]]
#positive
# extract 2
print(L[0][0][1])
# 4 through negative indexing
print(L[-2][1][1])

# Slicing
L = [1,2,3,4,5,6]

print(L[::-1])
print(L[0:3])
# in negative indexing last number will come first
print(L[-3:])
print(L[0::2])
print(L[-5:-2:2])
```

## Adding Items to a List

```python
# append -> it will add item as it is
L = [1,2,3,4,5]
L.append(True)
print(L)
```

```python
# extend -> it will add item in extended form
L = [1,2,3,4,5]
# if we append it it will append [6,7,8] as a list
L.extend([6,7,8])
print(L)
```

# Editing items in a List

```python
L = [1,2,3,4,5]

# editing with indexing
L[-1] = 500

# editing with slicing
L[1:4] = [200,300,400]

print(L)
```

## Deleting items from a List

```python
# del
L = [1,2,3,4,5]

# indexing
del L[-1]

# slicing -> # mutable(add/delete/edit)
del L[1:3]
print(L)
```

```python
# remove

L = [1,2,3,4,5]


# by value like if there is 100 or 5 then put 100 or 5

L.remove(5)
print(L)

L2=[1,2,3,100]

L2.remove(100)
print(L2)

```

```python
# pop
L = [1,2,3,4,5]

L.pop() # remove last item in list

print(L)
```

```python
# clear
L = [1,2,3,4,5]

L.clear() # empty the list

print(L)
```

## List Functions

```python
# len/min/max/sorted
L = [2,1,5,7,0]

print(len(L))
print(min(L))
print(max(L))
# function and not permanent
print(sorted(L,reverse=True))
```

```python
# count
L = [1,2,1,3,4,1,5]
L.count(1)
```

```python
# index
L = [1,2,1,3,4,1,5]
L.index(1)
```

```python
# reverse
L = [2,1,5,7,0]
# permanently reverses the list-> object
L.reverse()
print(L)
```

```python
# sort (vs sorted)
L = [2,1,5,7,0]
print(L)
print(sorted(L))
print(L)
# new memory in list
L.sort()
print(L)
```

```python
# copy -> shallow -> not effect the original list
# copy -> deep -> same as shallow but works on 2d/3d.. list
L = [2,1,5,7,0]
print(L)
print(id(L))
L1 = L.copy()
print(L1)
print(id(L1))
```

## List Comprehension

List Comprehension provides a concise way of creating lists.

newlist = [expression for item in iterable if condition == True]image.png

Advantages of List Comprehension
- More time-efficient and space-efficient than loops.
- Require fewer lines of code.
- Transforms iterative statement into a formula.

```python
# Add 1 to 10 numbers to a list
L = []

for i in range(1,11):
  L.append(i)

print(L)
```

```python
# i is expression,i is item range is iterable
[i for i in range(1,11)]
print(L)
```

```python
# scalar multiplication on a vector
# [-6,-9,-12] should be result
v = [2,3,4]
s = -3

L=[]
for i in v:
  L.append(i*s)

L

# i*s is expression and v is iterable
[i*s for i in v]
```

```python
# Add squares
L=[]
List1=[1,2,3,4,5]

for i in List1:
  L.append(i**2)

L

[i**2 for i in List1]
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50
L=[]
for i in range(1,51):
  if i%5==0:
    L.append(i)
L

# i is expression, iteration 1 to 50 and if condition
[i for i in range (1,51) if i%5==0]

```

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']
# s=[]
# for language in languages:
#  s.append(i.startswith('p'))

# print(s)

# i is expression, i iterable and if is condition
[i for i in languages if i.startswith('p')]
```

```python

basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']
# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

[fruit for fruit in my_fruits if (fruit in basket) and fruit.startswith('a') ]
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension

[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian products -> List comprehension on 2 lists together
# 1*5,1*6 and so on
L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i*j for i in L1 for j in L2]
```

## 2 ways to traverse a list

- itemwise
- indexwise

```python
# itemwise
L = [1,2,3,4]

for i in L:
  print(i)
```

```python
# index-wise
L = [1,2,3,4,6]

for i in range(0,len(L)):
  # index
  # print(i)
  # item-> corresponding values
  print(L[i])
```

## Zip

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
# Write a program to add items of 2 lists indexwise

L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

# pair the items
list(zip(L1,L2))

[i+j for i,j in zip(L1,L2)]
```
